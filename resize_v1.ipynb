```python
# This notebook uses images from 'comics-faces-paired' dataset 
# as the ones on which to apply transformations.

import os
import glob
import ntpath
from skimage import io, color
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def pad_image(image, target_shape):
    height, width = image.shape[:2]
    target_height, target_width = target_shape

    top_pad = (target_height - height) // 2
    bottom_pad = target_height - height - top_pad
    left_pad = (target_width - width) // 2
    right_pad = target_width - width - left_pad

    padded_image = cv.copyMakeBorder(image, top_pad, bottom_pad, left_pad, right_pad, cv.BORDER_CONSTANT, value=0)
    img = cv.resize(padded_image, dsize=(178, 218),
                    interpolation=cv.INTER_LINEAR)
    return img
```

```python
output_path = "/kaggle/working/comic-faces/train/"
os.makedirs(output_path + "comics/")
os.makedirs(output_path + "faces/")
```

```python
for img_path in glob.glob("/kaggle/input/comic-faces-paired-synthetic-v2/face2comics_v2.0.0_by_Sxela/face2comics_v2.0.0_by_Sxela/comics/*.jpg", recursive=True):
    if "samples" in img_path:
        print(img_path)
        continue
    img = io.imread(img_path)
    img = pad_image(img, (1308 + 218, 1068 + 178))
    base_name = ntpath.basename(img_path)
    io.imsave("/kaggle/working/comic-faces/train/comics/" + base_name, img)
```
